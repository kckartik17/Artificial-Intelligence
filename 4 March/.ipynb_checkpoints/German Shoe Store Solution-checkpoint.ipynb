```python
filter_usa = data_frame_women['Country'] == 'United States'
data_frame_women_usa = data_frame_women[filter_usa]
```

```python
data_frame_women_usa.groupby('Country').count() #match
```

```python
data_frame_women_20152016 = data_frame_women_usa[(data_frame_women_usa['Date']>=datetime.date(2015,1,1)) & (data_frame_women_usa['Date']<=datetime.date(2016,12,31))]
```

```python
data_frame_women_20152016.groupby('Country').count()
```

```python
df_women_size_month_usa = data_frame_women_20152016.groupby([data_frame_women_20152016['Date'].dt.strftime('%B'), 'Size (US)']).agg({'Size (US)': 'count'}).unstack(level=-1).transpose()
```

```python
df_women_size_month_usa
```

```python
df_women_size_month_usa.fillna(0, inplace=True)
```

```python
df_women_size_month_usa.columns.name = 'Months 2015-2016'
```

```python
df_women_size_month_usa.reset_index(level=0, drop=True, inplace=True)
```

```python
df_women_size_month_usa
```

```python
df_women_size_month_usa.sum() #dec 3 shoe are missing
```

```python
sns.clustermap(df_women_size_month_usa)
```

```python
def calculate_sem(row):
    return stats.sem(row)
```

```python
df_women_size_month_usa['SEM'] = df_women_size_month_usa.apply(calculate_sem, axis=1)
```

```python
df_women_size_month_usa
```

```python
def calculate_confidence_interval(row):
    return stats.t.interval(0.95, len(row)-3, loc=row[0:-2].mean(), scale=row['SEM'])
```

```python
df_women_size_month_usa['CONFIDENCE'] = df_women_size_month_usa.apply(calculate_confidence_interval, axis=1)
```

```python
df_women_size_month_usa
```

```python
counter = 0

while counter < len(df_women_size_month_usa):
    print(int(round(df_women_size_month_usa['CONFIDENCE'].iloc[counter][1])))
    counter+=1
```

```python
def inventory_prediction(xCI):
    return int(round(xCI[1], 0))

df_women_size_month_usa['PREDICTION'] = df_women_size_month_usa['CONFIDENCE'].apply(inventory_prediction)

sns.barplot(x=df_women_size_month_usa.index, y="PREDICTION", data=df_women_size_month_usa)
```
