```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
import datetime 
```

```python
data_frame = pd.read_excel(sheet_name='Al Bundy', io='German Shoe Store Data and problem stmt.xlsx')
```

```python
data_frame.columns = data_frame.iloc[2]
data_frame.columns 
```

```python
data_frame.set_index('InvoiceNo', inplace=True)
```

```python
data_frame.head()
```

```python
data_frame.dropna(inplace=True)
```

```python
data_frame.head()
```

```python
data_frame.columns.name = 'Features'
data_frame.columns = data_frame.columns.str.strip()
```

```python
data_frame.drop('Month', axis=1, inplace=True)
```

```python
data_frame.columns = data_frame.columns.fillna('delete')
```

```python
data_frame.drop('delete', axis=1, inplace=True)
```

```python
data_frame['Date'] = pd.to_datetime(data_frame['Date'])
```

```python
data_frame['UnitPrice'] = pd.to_numeric(data_frame['UnitPrice'], errors='ignore')
data_frame['Discount'] = pd.to_numeric(data_frame['Discount'], errors='ignore')
data_frame['SalePrice'] = pd.to_numeric(data_frame['SalePrice'], errors='ignore')
```

```python
filter_men = data_frame['Gender'] == 'Male'
data_frame_men = data_frame[filter_men]
```

```python
data_frame_men.head()
```

```python
df_sales_by_country_size = data_frame_men.groupby(['Size (US)', 'Country']).agg({'Country': 'count'}).unstack(level=-1)
```

```python
df_sales_by_country_size
```

```python
df_sales_by_country_size.sum(axis=0) #matched
```

```python
filter_usa = data_frame_men['Country'] == 'United States'
data_frame_men_usa = data_frame_men[filter_usa]
```

```python
data_frame_men_usa.groupby('Country').count() #match
```

```python
data_frame_men_20152016 = data_frame_men_usa[(data_frame_men_usa['Date']>=pd.Timestamp(2015,1,1)) & (data_frame_men_usa['Date']<=pd.Timestamp(2016,12,31))]
```

```python
data_frame_men_20152016.groupby('Country').count()
```

```python
df_men_size_month_usa = data_frame_men_20152016.groupby([data_frame_men_20152016['Date'].dt.strftime('%B'), 'Size (US)']).agg({'Size (US)': 'count'}).unstack(level=-1).transpose()
```

```python
df_men_size_month_usa
```

```python
df_men_size_month_usa.fillna(0, inplace=True)
```

```python
df_men_size_month_usa.columns.name = 'Months 2015-2016'
```

```python
df_men_size_month_usa.reset_index(level=0, drop=True, inplace=True)
```

```python
df_men_size_month_usa
```

```python
df_men_size_month_usa.sum()
```

```python
sns.clustermap(df_men_size_month_usa)
```

```python
def calculate_sem(row):
    return stats.sem(row)
```

```python
df_men_size_month_usa['SEM'] = df_men_size_month_usa.apply(calculate_sem, axis=1)
```

```python
df_men_size_month_usa
```

```python
def calculate_confidence_interval(row):
    return stats.t.interval(0.95, len(row)-3, loc=row[0:-2].mean(), scale=row['SEM'])
```

```python
df_men_size_month_usa['CONFIDENCE'] = df_men_size_month_usa.apply(calculate_confidence_interval, axis=1)
```

```python
df_men_size_month_usa
```

```python
counter = 0

while counter < len(df_men_size_month_usa):
    print(int(round(df_men_size_month_usa['CONFIDENCE'].iloc[counter][1])))
    counter+=1
```

```python
def inventory_prediction(xCI):
    return int(round(xCI[1], 0))

df_men_size_month_usa['PREDICTION'] = df_men_size_month_usa['CONFIDENCE'].apply(inventory_prediction)

sns.barplot(x=df_men_size_month_usa.index, y="PREDICTION", data=df_men_size_month_usa)
```
